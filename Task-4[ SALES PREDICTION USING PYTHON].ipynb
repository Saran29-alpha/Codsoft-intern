```python
# importing requirede libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
## loading dataset

sales_data=pd.read_csv("advertising.csv")
sales_data.head(10)
```

```python
sales_data.isnull().sum()
```

```python
sales_data.dtypes
```

```python
sales_data.describe()
```

```python
#visualize the given data

sns.pairplot(sales_data,x_vars=["TV","Radio","Newspaper"],y_vars="Sales",height=6,aspect=0.7)
plt.show()
```

```python
x=sales_data[["TV","Radio","Newspaper"]]
y=sales_data["Sales"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
# evaluate the model 

print("MSE",mean_squared_error(y_test,y_pred))
print("R2 Score",r2_score(y_test,y_pred))
```

```python
## Prediction

### using python data structures

data_sample={
    "TV":[30,40,228,33],
    "Radio":[20,40,10,30],
    "Newspaper":[25,5,90,15]}
data_samples=pd.DataFrame(data_sample)
data_sample_pred=model.predict(data_samples)
for i,prediction in enumerate(data_sample_pred,start=1):
    print(f"Sample {i}: sales_data = {prediction:.2f}")
```

```python
data_sample_2=[22.9,33,27.3]
reshapes_data_sample_2=np.array(data_sample_2).reshape(1,-1)
data_samples_2=model.predict(reshapes_data_sample_2)
print("Sales prediction:",data_samples_2)
```
